creating a series from dictionary

```python
li = pd.Series({'a':10, 'b':20, 'c':30, 'd':40, 'e':50 })
li
```

```python
# to change the index position of the dictionary
li = pd.Series({'a':10, 'b':20, 'c':30, 'd':40 }, index=['e', 'd', 'c', 'b', 'a'])
li
```

Extracting Individual element

```python
# extracting a single element
s1 = pd.Series([1,2,3,4,5,6])
s1[3]
```

```python
# extracting element from start and stopping at the specified index
s1 = pd.Series([1,2,3,4,5,6,7,8])
s1[:5]
```

```python
# extracting element from the back
s1 = pd.Series([1,2,3,4,5,6])
s1[-3:]
```

Basic math operation on series

```python
# Adding a scalar value to series element
s1 = pd.Series([1,2,3,4,5,6])
s1 + 10
```

```python
# adding two series
s1 = pd.Series([1,2,3,4,5,6])
s2 = pd.Series([5,10,15,20,25,30])
s1 + s2
```

Dataframe In-Built function

```python
# head() is used to check the first five rows of a datframe
# shape() used to know how many rows and columns in the datframe
# tail() to show the last five rows 
# describe() basic description

import pandas as pd
iris = pd.read_csv('2018_Central_Park_Squirrel_Census_Data.csv')
iris
```

```python
iris.describe()
```

```python
iris.tail()
```

```python
iris.head()
```

```python
#extracting from datafram use .iloc[]
iris.iloc[2:6,0:5]
```

```python
# using the .loc[] method
iris.loc[0:5, ('Unique Squirrel ID', 'Hectare', 'Shift')]
```

```python
# Dropping colums
# axis will equal to 1 to drop column and equal to 0 top drop row
# add to list to drop multiple colums
iris.drop(['Shift', 'Date'],  axis=1)
```

```python
#dropping rows
iris.drop([1,4,5], axis=0)
```

```python
#to use other pandas function just invoke the method name instead of using the describe method
# iris.mean()
# iris.median()
# iris.max()
# iris.min()
# iris.std()

# applying simple function to multiple data using the apply()
def half(s):
    return s*0.5
iris[['X', 'Y']].apply(half)
```

```python
def double_value(value):
    return value*2
iris[['X', 'Y']].apply(double_value)
```

```python
# value count function in the dataframe using value_count()
iris['Primary_Fur_Color'].value_counts()
```

```python
# sorting data based on index value using sort_values() it sorts from minimum to maximum
iris.sort_values(by='Hectare Squirrel Number')
```
